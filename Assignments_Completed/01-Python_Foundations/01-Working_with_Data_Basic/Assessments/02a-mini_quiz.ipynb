### 1️⃣ Quiz Question - Given the python list: 

```python
L = [13,17,31,37,73]
```
Write one line of Python code that removes the value 31 from the list by value, not by index.
After your code executes, the list should become:

```python
L = [13,17,31,37,73]
```

````python
# Answer 1

# I needed to declare it and I threw in a print statement for clairty
L = [13,17,31,37,73]
print(f"Before: {L}")

# One line of code:
L.remove(31)

print(f"After: {L}")
````
