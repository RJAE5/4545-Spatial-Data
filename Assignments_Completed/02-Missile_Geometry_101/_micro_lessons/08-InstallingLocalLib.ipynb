# Installing Our Own Local Library

Why do we need to do this? Well, coming up in the next set of lessons, were going to use our own functions. These functions are provided by Griffin, obtained from multiple sources and packaged together for convenience. These functions allow us to do geospatial comparisons, projections, interpolations and more, but we want to keep those functions organized in their own file (module). We also want to keep this project organized as well. In order to do this we can actually install our own library (module) locally, and not permanently (for now). The geospatial module lives in `src/wdo` where w.d.o = World Defense Organization and the rest of this cell will describe how to install it for use.

## How to Install Our Own Lib

Our project contains **reusable Python code** that lives in this folder:

```
src/wdo
```

Instead of placing Python modules directly in the project root, we use a **`src` layout** to keep the code organized and separate from things like:

- `notebooks`
- `datasets`
- `markdown files`
- `setup scripts`

**Example structure:**

```
.
├── data
├── notebooks
├── pyproject.toml
└── src
    └── wdo
```

Notice the `pyproject.toml` file. This is important.

---

### What `pyproject.toml` Does

The `pyproject.toml` file tells Python's packaging tools:

1. The project name (`wdo`)
2. Which Python version is required
3. **Where the package code lives (`src/`)**

So Python knows that the real code is here:

```
src/wdo
```

---

### What This Command Does

Before running any notebooks or scripts, we run:

```bash
pip install -e .
```

This command:

1. Installs the project into the current virtual environment  
2. Makes the `wdo` package importable anywhere in the project  
3. Uses **editable mode**, meaning changes to the code are immediately available

---

### Why Editable Mode Matters

After running:

```bash
pip install -e .
```

we can write:

```python
from wdo_geo import point_in_polygon
```

and Python will find the code inside:

```
src/wdo
```

Even better, if we modify files in `src/wdo`, we **do not need to reinstall** the package.

---

# Quick Summary

Run this once after you clone or copy the files into your own project:

This tells Python:

## Problem Three

- Using GeoJson.IO as a helper
  - Create a polygon that is approx same size as a small state.
  - Create two points where one is inside the polygon and one is outside.
- Run the point_in_polygon function to show your polygons and points are accurate and that you can use the wdo.geo library.
  - `def point_in_polygon(point: LatLon, polygon: List[LatLon]) -> bool`

````python

from collections import namedtuple
from wdo.geo import point_in_polygon

# Define LatLon structure
LatLon = namedtuple("LatLon", ["lat", "lon"])

# Create a polygon roughly the size of a small state (simple rectangle)
polygon = [
    LatLon(30.0, -100.0),  
    LatLon(30.0, -95.0),   
    LatLon(35.0, -95.0),  
    LatLon(35.0, -100.0) 
]

# Create test points
inside_point = LatLon(32.5, -97.0) 
outside_point = LatLon(36.0, -97.0)

# Run point_in_polygon
inside_result = point_in_polygon(inside_point, polygon)
outside_result = point_in_polygon(outside_point, polygon)

# Print results
print("Inside point:", inside_point, "->", inside_result)
print("Outside point:", outside_point, "->", outside_result)
````
