# 📘 01 - Scalar Types

Review and practice working with Python's basic scalar data types.

## ✅ Learning Goals
- Recognize scalar types (`int`, `float`, `str`, `bool`)
- Use the `type()` function to identify types
- Use variables and assignment

````python
# 🔧 Define some variables
num = 42
pi = 3.14159
name = 'Python'
is_cool = True
````

````python
# 🔍 Check their types
print(type(num))
print(type(pi))
print(type(name))
print(type(is_cool))
````
