# 📘 02 - Python Tuples

Learn what tuples are and how to use them.

## ✅ Learning Goals
- Create and access tuples
- Understand immutability
- Use tuple unpacking

````python
# 🔧 Create a Tuple
coords = (5, 10)
print(coords)
````

````python
# 🔍 Indexing
print('X:', coords[0])
print('Y:', coords[1])
````

````python
# 🔁 Tuple Unpacking
x, y = coords
print('x + y =', x + y)
````
