## Q4. File I/O
Write Python code that saves `Hello World` to a file called `hello.txt`, then reads it back and prints the contents.

```python
with open("hello.txt", "w") as f:
    f.write("Hello World")
with open("hello.txt", "r") as f:
    print(f.read())
```

## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute y = x^2
y = x**2

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.title('Line Plot of y = x^2')

# Show grid
plt.grid(True)

# Display plot
plt.show()
```

## Q9. Timing
Compare the runtime of squaring numbers using a Python loop vs. NumPy. Use `%timeit` for each.

```python
n = 1_000_000

def loop_square():
    result = []
    for i in range(n):
        result.append(i**2)
    return result

%timeit loop_square()

import numpy as np

arr = np.arange(n)

%timeit arr**2
```
