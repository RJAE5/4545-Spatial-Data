```python

import pandas as pd
import numpy as np
import re

df = pd.read_csv("./../../data/survey.csv")

print(df.head())
print(df.info())

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].str.strip()

df = df.replace(r"^\s*$", np.nan, regex=True)

print(df.isnull().sum())

if "gender" in df.columns:
    df["gender"] = df["gender"].str.lower().replace({
        "m": "Male", "male": "Male",
        "f": "Female", "female": "Female",
        "other": "Other", "non-binary": "Other"
    })

if "exercise_freq" in df.columns:
    df["exercise_freq"] = df["exercise_freq"].str.lower().replace({
        "never": "Never",
        "rarely": "Rarely",
        "sometimes": "Sometimes",
        "often": "Often",
        "daily": "Daily"
    })

for col in ["age", "height_cm", "weight_kg", "satisfaction"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.extract(r"(\d+\.?\d*)")[0]
        df[col] = pd.to_numeric(df[col], errors="coerce")

if "submitted_at" in df.columns:
    df["submitted_at"] = pd.to_datetime(df["submitted_at"], errors="coerce")

if "email" in df.columns:
    email_pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    df = df[df["email"].astype(str).str.match(email_pattern, na=False)]

if "consent_given" in df.columns:
    df["consent_given"] = df["consent_given"].astype(str).str.lower().map({
        "yes": True, "true": True, "1": True,
        "no": False, "false": False, "0": False
    })

df = df.drop_duplicates()

df.to_csv("survey_clean.csv", index=False)

print(df.info())
print(df.head())
```
